Project Title: World Population Growth Visualization with Matplotlib

1. Top 10 Most Populated Countries (2022)
Use a bar plot.

X-axis: Country | Y-axis: Population in billions.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\jumde\OneDrive\Desktop\world_population.csv")
```

```python
df.head(3)
```

```python
x=df[["Country/Territory","2022 Population"]]
x
```

```python
df2=x.sort_values("2022 Population",ascending=False).head(10)
df2
```

```python
x=df2["Country/Territory"]
y=df2["2022 Population"]
```

```python
plt.bar(x,y)
plt.xlabel("Countries")
plt.ylabel("Population_2022")
plt.show()
```

2. Population Distribution by Continent (2022)
Pie chart showing the percentage share of each continent.


```python
p = df[["Country/Territory", "2022 Population"]].sort_values("2022 Population", ascending=False).head(10)
```

```python
plt.pie(p["2022 Population"],labels=p["Country/Territory"])
plt.show()
```

3. Scatter Plot: Country vs Density
Show relationship between population density and total population.


```python
m=df["Density (per km²)"]
df["Total_Population"]=df["2022 Population"]+df["2015 Population"]+df["2020 Population"]+df["2010 Population"]+df["2000 Population"]+df["1990 Population"]+df["1980 Population"]+df["1970 Population"]
n=df["Total_Population"]
```

```python
plt.scatter(m,n)
plt.xlabel("Density")
plt.ylabel("Total Population")
plt.show()
```

4. Subplots
Combine 2–3 small charts (bar + pie + line + scatter) in one figure using plt.subplot().

```python
plt.subplot(1,3,1)
plt.pie(p["2022 Population"],labels=p["Country/Territory"])
plt.subplot(1,3,2)
plt.bar(x,y)
plt.xlabel("Countries")
plt.ylabel("Population_2022")
plt.subplot(1,3,3)
plt.scatter(m,n)
plt.xlabel("Density")
plt.ylabel("Total Population")
plt.figure(figsize=(100, 500))
plt.tight_layout()
plt.show()
```
